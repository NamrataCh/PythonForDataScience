```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("data/HollywoodMovies.csv")
df.info()
```

#### 1.Find the highest-rated movie in the “Quest” story type.

```python
highest_rating=df[df["Story"]=="Quest"]["RottenTomatoes"].max()
high_rated_movie=df[df["RottenTomatoes"]==highest_rating]
print(high_rated_movie)
```

#### 6.1: From the raw data below create a Pandas Series

   - a) Print all elements in lower case
   - b) Print all the elements in upper case

#### 6.2: From the raw data below create a Pandas Series' Atul', 'John ', ' jack ', 'Sam'
   - a) Print all elements after stripping spaces from the left and right
   - b) Print all the elements after removing spaces from the left only
   - c) Print all the elements after removing spaces from the right only

```python
series2=pd.Series([' Atul', 'John ', ' jack ', 'Sam'])
series2
```

```python
#a) Print all elements after stripping spaces from the left and right
print(series2.str.strip().tolist())
print("_______"*5)
#b) Print all the elements after removing spaces from the left only
print(series2.str.lstrip().tolist())
print("_______"*5)
#c) Print all the elements after removing spaces from the right only
print(series2.str.rstrip().tolist())
print("_______"*5)
```

#### 6.3: -Create a series from the raw data below 'India_is_big', 'Population_is_huge', np.nan, 'Has_diverse_culture'
   - a)split the individual strings wherever ‘_’ comes and create a list out of it.
   - b)Access the individual elementsof a list
   - c)Expand the elements so that all individual elements get splitby ‘_’ and insteadof list returns individual elements

```python
series3=pd.Series(['India_is_big', 'Population_is_huge', np.nan, 'Has_diverse_culture'])
series3
```

```python
#a)split the individual strings wherever ‘_’ comes and create a list out of it.
print(series3.str.split("_"))
print("_______"*5)

#b)Access the individual elementsof a list
ind_ele_list=series3.str.split("_").tolist()
print(ind_ele_list)
print("_______"*5)

#c)Expand the elements so that all individual elements get splitby ‘_’ and insteadof list returns individual elements
all_elems=[]
for item in ind_ele_list:
    if isinstance(item, list): 
        for value in item:            
            all_elems.append(value)
        
    else:
        all_elems.append(item)
print(all_elems)
```
