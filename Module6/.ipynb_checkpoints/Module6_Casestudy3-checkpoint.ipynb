1. Compute -- Use seaborn to create a jointplot to compare the Time on Website and Yearly Amount Spent columns. Is there a correlation?



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
df=pd.read_csv("data/FyntraCustomerData.csv")
```

```python
df.info()
```

```python
# Use seaborn to create a jointplot to compare the Time on Website and Yearly Amount Spent columns. Is there a correlation?
df_Time_Web_Amnt_Spent=df[["Time_on_Website", "Yearly_Amount_Spent"]]
sns.jointplot(x="Time_on_Website", y="Yearly_Amount_Spent",  data=df_Time_Web_Amnt_Spent)
```

```python
# Is there a correlation? : Both have strong correlation
sns.heatmap(df_Time_Web_Amnt_Spent.corr())
```

#### 2. Compute – Do the same as above but now with Time on App and Yearly Amount Spent. Is this correlation stronger than 1st One?

```python
df_Time_App_Amnt_Spent=df[["Time_on_App", "Yearly_Amount_Spent"]]
sns.jointplot(x="Time_on_App", y="Yearly_Amount_Spent", kind="scatter", data=df_Time_App_Amnt_Spent)
```

```python
# Is this correlation stronger than 1st One? : yes, it's  stronger than first one
sns.heatmap(df_Time_App_Amnt_Spent.corr())
```

#### 5. Compute – Train and Test the data and answer multiple questions -- What is the use of random_state=85?

```python
X=df[['Avg_Session_Length', 'Time_on_App', 'Time_on_Website', 'Length_of_Membership']]
y=df['Yearly_Amount_Spent']
X_train,X_test, y_train, y_test=train_test_split(X,y, test_size=0.20, random_state=85)
```

#### what is the use of random_state? 
Random_state is used to set the seed for the random generator so that we can ensure that the results that we get can be reproduced. Because of the nature of splitting the data in train and test is randomised you would get different data assigned to the train and test data unless you can control for the random factor.

#### 6. Compute – Predict the data and do a scatter plot. Check if actual and predicted data match.

```python
model=LinearRegression()
model.fit(X_train, y_train)
y_predict=model.predict(X_test)
```

```python
# check if actual and predicted data match
plt.scatter(x=y_predict, y=y_test, )
plt.xlabel("Y-Predict")
plt.ylabel("Y-Test")
plt.show()
```

#### 7. What is the value of Root Mean Squared Error?

```python
mse=mean_squared_error(y_test, y_predict)
print(sqrt(mse))
```
