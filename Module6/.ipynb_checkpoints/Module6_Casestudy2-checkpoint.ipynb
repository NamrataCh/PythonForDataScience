1. Load the data from “cereal.csv” and plot histograms of sugar and vitamin content across different cereals.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data/cereal.csv")
df.head()
```

```python
sugar_data=df["sugars"]
vitamin_data=df["vitamins"]

plt.hist([sugar_data, vitamin_data], color=["orange", "green"])
plt.title("Sugars and Vitamins")

plt.xlabel("Sugars and Vitamins")
plt.legend(["Sugars", "Vitamins"])

plt.show()
```

### 2. The names of the manufacturers are coded using alphabets, create a new column with their full name using the below mapping.
    'N': 'Nabisco',
    'Q': 'Quaker Oats',
    'K': 'Kelloggs',
    'R': 'Raslston Purina',
    'G': 'General Mills',
    'P' :'Post' ,
    'A':'American Home Foods Products
    
Create a bar plot where each manufacturer is on the y axis and the height of the bars depicts the number of cereals manufactured by them.

[Hint: Try using count plot this time or bar method of pyplot]


```python
dict_mfr={
    'N': 'Nabisco',
    'Q': 'Quaker Oats',
    'K': 'Kelloggs',
    'R': 'Raslston Purina',
    'G': 'General Mills',
    'P': 'Post',
    'A': 'American Home Foods Products'
}

df["manufacturer"]=[dict_mfr[mfr] for mfr in df['mfr']]
df.head()
```

```python
mfr_grouping=df.groupby(["manufacturer"], as_index=False).count()
mfr_grouping
```

```python
mfr_grouping.columns
```

```python
x=mfr_grouping["manufacturer"]
y=mfr_grouping["mfr"]

plt.bar(x,y)
plt.setp(plt.gca().get_xticklabels(), rotation=90, horizontalalignment='right')
plt.xlabel("Manufacturer")
plt.ylabel("Count")
plt.title("number of cereals manufactured by Manufactures")
plt.show()
```

#### 3. Extract the rating as your target variable ‘y’ and all numerical parameters as your predictors ‘x’. Separate 25% of your data as a test set.

```python
y=df["rating"]
x=df.iloc[:, 3:15]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=.25, random_state=42)
```

### 4. Fit a linear regression module and measure the mean squared error on the test dataset.

[ Hint: Explore linear models and metrics section of sklearn documentation]

```python
from sklearn.linear_model import LinearRegression
import numpy as np

model=LinearRegression()
model.fit(X_train, y_train)

y_predict=model.predict(X_test)

plt.scatter(np.array(y_predict), np.array(y_test))
plt.show()
```
