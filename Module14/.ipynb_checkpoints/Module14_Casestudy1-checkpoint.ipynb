```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("data/bio-degradabale-data.csv", sep=";")
df.head()
```

```python
df.shape
```

```python
df.columns=range(42)
df.head()
```

```python
X=df.iloc[:, 0:41]
y=df[41]
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.20, random_state=42)
```

```python
# implement Logistic Regression classifier
log_model=LogisticRegression()
log_model.fit(X_train, y_train)
y_predict=log_model.predict(X_test)
```

```python
print(accuracy_score(y_test, y_predict))
```

```python
# implement randomforest classifier to improve the accuracy score
random_model=RandomForestClassifier()
random_model.fit(X_train,y_train)
y_predict_rand=random_model.predict(X_test)
```

```python
print(accuracy_score(y_test, y_predict_rand))
```
