```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import mean_squared_error
```

```python
# 1. Load the file “zoo.data” and look at the info and the first five rows. 
# The first column denotes the animal name and the last one specifies a high-level class for the corresponding animal.
data=pd.read_csv("data/zoo.csv")
data.head()
```

```python
# 2. Find out the unique number of high-level cclasses
unique_class=data["class_type"].value_counts()
unique_class
```

```python
# 3. Use the 16-intermediate feature and perform an agglomerative clustering.

agg_cluster=AgglomerativeClustering(n_clusters=4)
agg_predict=agg_cluster.fit_predict(data.iloc[:, 1:17])

```

```python
# 4. Compute the mean squared error by comparing the actual class and predicted high-level class.
print(mean_squared_error(data["class_type"],agg_predict ))
```
