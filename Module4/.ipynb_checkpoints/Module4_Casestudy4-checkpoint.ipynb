#### Approach to Solve

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("data/BigMartSalesData.csv")
df.head()
```

1. Plot Total Sales Per Month for the Year 2011. How the total sales have increased over months in the Year 2011. Which month has the lowest Sales?

```python
df_grouped=df.query("Year==2011").filter(["Month", "Amount"]).groupby("Month", as_index=False).sum()
plt.plot(df_grouped["Month"],df_grouped["Amount"])
plt.title("Total Sales Per Month for the Year 2011")
plt.xlabel("Month")
plt.ylabel("Total Sales per Month")
plt.show()
```

#### 2. Plot Total Sales Per Month for the Year 2011 as Bar Chart. Is Bar Chart Better to visualize than a Simple Plot?

```python
df_grouped=df.query("Year==2011").filter(["Month", "Amount"]).groupby("Month", as_index=False).sum()
plt.bar(df_grouped["Month"],df_grouped["Amount"])
plt.title("Total Sales Per Month for the Year 2011")
plt.xlabel("Month")
plt.ylabel("Total Sales per Month")
plt.show()
```

#### 3. Plot Pie Chart for the Year 2011 Country Wise. Which Country contributes the highest towards sales?

```python
df_country_grouped=df.query("Year==2011").filter(["Country", "Amount"]).groupby("Country", as_index=False).sum()
plt.figure(figsize =(10, 7))


plt.pie(df_country_grouped["Amount"], labels=df_country_grouped["Country"], autopct='%1.1f%%', shadow=True)

plt.show()
```

#### 4. Plot Scatter Plot for the invoice amounts and see the concentration of the amount. In which range most of the invoice amounts are concentrated 

```python
df_conc_amnt=df.filter(["InvoiceDate", "Amount"]).groupby("InvoiceDate", as_index=False).sum()

plt.scatter(df_conc_amnt["InvoiceDate"], df_conc_amnt["Amount"])
plt.ylim(0, 100000)
plt.xlabel("Invoice Date")
plt.ylabel("Amount")

# range where most of the invoice amounts are concentrated: 20000 - 40000
```

#### Enhancements for code
You can try these enhancements in code
1. Change the bar chart to show the value of the bar
2. In Pie Chart Play With Parameters shadow=True, startangle=90, and see how different the chart looks
3. In scatter plot change the color of Scatter Points

```python
# 1. Change the bar chart to show the value of the bar
import numpy as np
colors=['blue','green','red','orange','cyan','black','pink','magenta']

df_grouped=df.query("Year==2011").filter(["Month", "Amount"]).groupby("Month", as_index=False).sum()
plt.bar(df_grouped["Month"],df_grouped["Amount"], color=colors)
plt.title("Total Sales Per Month for the Year 2011")
plt.xlabel("Month")
plt.ylabel("Total Sales per Month")
plt.show()
```
