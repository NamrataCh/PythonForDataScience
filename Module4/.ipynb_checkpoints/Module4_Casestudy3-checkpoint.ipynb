1. You are given a dataset, which is present in the LMS, containing the number of hurricanes occurring in the United States along the coast of the Atlantic. Load the data from the dataset into your program and plot a Bar Graph of the data, taking the Year as the x-axis and the number of hurricanes occurring as the Y-axis.




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data/Hurricanes.csv")
df.head()
```

```python
plt.bar(df['Year'], df["Hurricanes"])
plt.xlabel("Year")
plt.ylabel("Hurricanes")
plt.show()
```

#### 2. The dataset given, records data of city temperatures over the years 2014 and 2015. Plot the histogram of the temperatures over this period for the cities of San Francisco and Moscow.

```python
df_temprature=pd.read_csv("data/CityTemps.csv")
plt.hist(df_temprature['San Francisco'], len(df_temprature['San Francisco']), alpha=0.5)
plt.hist(df_temprature['Moscow'], len(df_temprature['Moscow']), alpha=0.5)
plt.xlabel("Temprature")
plt.ylabel("Frequency")
plt.show()
```

```python
X = [1,2,3,4]
y = [20, 21, 20.5, 20.8]
```

```python
# 5.1: Draw a Simple plot
plt.plot(X,y)
plt.show()
```

```python
#5.2: Configure the line and markers in a simple plot
plt.plot(X, y, linewidth=1,marker="+")

plt.show()
```

```python
# 5.3: configure the axes
plt.plot(X,y)
plt.xlabel('X-axis Label', fontweight='bold', color='orange',
   fontsize='17', horizontalalignment='center')

plt.show()
```

```python
#5.4: Give the title of Graph & labels of the x axis and y axis
plt.plot(X,y)
plt.title("Graph Title")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
#5.5: Give error bar if y_error = [0.12, 0.13, 0.2, 0.1]
y_error = [0.12, 0.13, 0.2, 0.1]
plt.plot(X,y,y_error)
```

```python
# 5.9: Create a dataframe from the following data, Draw a Scatterplot of preTestScore and postTestScore, with the size of each point
#determined by age

data={
'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
'female': [0, 1, 1, 0, 1],
'age': [42, 52, 36, 24, 73],
'preTestScore': [4, 24, 31, 2, 3],
'postTestScore': [25, 94, 57, 62, 70]
}

df_data=pd.DataFrame(data)

plt.scatter(df_data["preTestScore"], df_data["postTestScore"], s=df_data['age'])
plt.xlabel("preTestScore")
plt.ylabel("postTestScore")
plt.show()
```

```python
#5.10: Draw a Scatterplot from the data in question 9 of preTestScore and
# postTestScore with the size = 300 and the color determined by sex

plt.scatter(df_data["preTestScore"], df_data["postTestScore"], s=300, c=df_data['female'])
plt.xlabel("preTestScore")
plt.ylabel("postTestScore")
plt.show()
```
