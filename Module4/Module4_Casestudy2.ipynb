#### Approach to Solve
You have to use the fundamentals of Numpy and Pandas covered in module 4.
1. Read the three CSV files which contain the score of the same students in term1 of each Subject
2. Remove the name and ethnicity column (to ensure confidentiality)
3. Fill missing score data with zero

```python
import pandas as pd
import numpy as np
```

```python
#1. Read the three CSV files which contain the score of the same students in term1 of each Subject

df_math_score=pd.read_csv("data/MathScoreTerm1.csv")
df_ds_score=pd.read_csv("data/DSScoreTerm1.csv")
df_physics_score=pd.read_csv("data/PhysicsScoreTerm1.csv")
```

```python
print("________________Maths Score___________________________")
print(df_math_score.head())
print("________________DS Score_______________________________________")
print(df_ds_score.head())
print("_________________Physics Score_________________________________")
print(df_physics_score.head())
```

```python
#2. Remove the name and ethnicity column (to ensure confidentiality)
cols_to_drop=['Name', 'Ethinicity']

df_math_score.drop(columns=cols_to_drop, inplace=True)
df_ds_score.drop(columns=cols_to_drop, inplace=True)
df_physics_score.drop(columns=cols_to_drop, inplace=True)
```

```python
print("________________Maths Score___________________________")
print(df_math_score.head())
print("________________DS Score_______________________________________")
print(df_ds_score.head())
print("_________________Physics Score_________________________________")
print(df_physics_score.head())
```

```python
#3. Fill missing score data with zero
df_math_score.fillna(0)
df_ds_score.fillna(0)
df_physics_score.fillna(0)
```

```python
#4. Merge the three files
merger_df=df_math_score.merge(df_ds_score, on="ID", suffixes=("_math", "_ds")).merge(df_physics_score, on="ID")
result_df=merger_df.filter(["ID", "Age", "Subject", "Sex", "Score", "Score_math", "Score_ds"])
result_df.rename(columns={"Score":"Score_physics"})
result_df.head()
```

```python
#5. Change Sex(M/F) Column to 1/2 for further analysis
result_df['Sex']=result_df['Sex'].map({'M':1, 'F':2})
result_df.head()
```

```python
# 6. Store the data in a new file – ScoreFinal.csv
result_df.to_csv("data/ScoreFinal.csv")
```
